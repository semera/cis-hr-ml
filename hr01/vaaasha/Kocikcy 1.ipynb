Using a classifier find the expected _cat_ generated by the system based on input data. 

```python
ftraining = r"C:\P\Machine Learning\kocicky 1\learning.csv"
import numpy as np
training = np.loadtxt(ftraining,delimiter=',')
training.shape
```

Training set contains input parameters (machine set up) in the first 3 column and the last one is resulted cat 1/0

```python
X_training = training[:,:-1]
y_training = training[:,-1]
print(X_training[0:2],y_training[0:2])
```

Scale the X to mean = 0 using **StandardScaler**, since MLPClassifier below likes prescalled data

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_training)
X_training_scaled = scaler.transform(X_training)
print(X_training_scaled[0:2])
```

Train the neural network using **MLPClassifier**

```python
from sklearn.neural_network import MLPClassifier
clf = MLPClassifier(solver='lbfgs', alpha=1e-5,hidden_layer_sizes=(12,6), random_state=1,verbose=True)
clf.fit(X_training_scaled, y_training)
print(clf.score(X_training_scaled, y_training))
```

Predict using testing data. Don't forget to scale the input too using StandardScaler

```python
ftraining = r"C:\P\Machine Learning\kocicky 1\test1000.csv"
testing = np.loadtxt(ftraining,delimiter=',')
X_testing = scaler.transform(testing)
y_testing = clf.predict(X_testing)
```

Output to the text file \n separated

```python
np.savetxt(r"C:\P\Machine Learning\kocicky 1\output.csv",y_testing,delimiter='\n',fmt='%d')
```

```python
clf.predict_proba(X_testing)
```
